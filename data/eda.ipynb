```python
def calc_spread(odd_home: float, odd_draw: float, odd_away: float) -> float:
    """
    Calculate the bookmaker spread (overround) given three decimal odds.

    Parameters
    ----------
    odd_home : float
        Decimal odds for the home win.
    odd_draw : float
        Decimal odds for the draw.
    odd_away : float
        Decimal odds for the away win.

    Returns
    -------
    float
        Spread (overround) as a percentage, e.g. 7.2 means 7.2%.
    """
    total_prob = (1/odd_home) + (1/odd_draw) + (1/odd_away)
    spread_pct = (total_prob - 1) * 100
    return round(spread_pct, 2)



```
